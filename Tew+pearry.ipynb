3.Data Preparation

```python
#encode categorical variables 
gender_enc = {val:ii for ii, val in enumerate(np.unique(df['Gender']))}
df['Gender'] = np.where(df['Gender'].isin(gender_enc), df['Gender'].map(gender_enc), np.nan)

smoke_enc = {val:ii for ii, val in enumerate(np.unique(df['SMOKE']))}  
df['SMOKE'] = np.where(df['SMOKE'].isin(smoke_enc), df['SMOKE'].map(smoke_enc), np.nan)

nobeysdad_enc = {val:ii for ii, val in enumerate(np.unique(df['NObeyesdad']))}
df['NObeyesdad'] = np.where(df['NObeyesdad'].isin(nobeysdad_enc), df['NObeyesdad'].map(nobeysdad_enc), np.nan)

df
```

```python
#Normalize numeric column into 2 decimal places(age,weight,height)
df['Age'] = df['Age'].apply(lambda x: round(x, 2))
df['Weight'] = df['Weight'].apply(lambda x: round(x, 2))
df['Height'] = df['Height'].apply(lambda x: round(x, 2))
df
```

Comparison of eating habit by gender ( Using six radar)

```python
df.replace({'yes': 1, 'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always':3,'Public_Transportation': 0, 'Walking': 1,'Automobile':2,'Motorbike':3,'Bike':4}, inplace=True)
eating_habits = ['FAVC', 'FCVC', 'NCP', 'CAEC', 'CH2O', 'CALC']
eating_data = df[eating_habits]

# Radar chart (requires normalization)
normalized_data = (eating_data - eating_data.mean()) / eating_data.std()
#normalized_data = normalized_data.head(10)  # Considering only the first 10 persons for visualization

labels=np.array(eating_habits)
stats=normalized_data.iloc[0].values

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()

stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='red', alpha=0.25)
ax.plot(angles, stats, color='red', linewidth=2)

stats=normalized_data.iloc[1].values
stats=np.concatenate((stats,[stats[0]]))
ax.fill(angles, stats, color='blue', alpha=0.25)
ax.plot(angles, stats, color='blue', linewidth=2)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)
plt.title('Eating Habits Radar Chart')
plt.show()
```

Comparison of eating habit between genders (Using distribution)

```python
attributes = ['FAVC', 'FCVC', 'NCP', 'CAEC','CALC', 'Gender']

# Melt the dataframe to long format for easy plotting
melted_data = pd.melt(df[attributes], id_vars=['Gender'], var_name='Eating Habit')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Eating Habit', y='value', hue='Gender', data=melted_data)
plt.title('Comparison of Eating Habits Between Genders')
plt.xlabel('Eating Habit')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```
