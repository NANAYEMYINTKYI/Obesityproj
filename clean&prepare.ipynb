3.Data Preparation

```python
#encode categorical variables 
gender_enc = {val:ii for ii, val in enumerate(np.unique(df['Gender']))}
df['Gender'] = np.where(df['Gender'].isin(gender_enc), df['Gender'].map(gender_enc), np.nan)

smoke_enc = {val:ii for ii, val in enumerate(np.unique(df['SMOKE']))}  
df['SMOKE'] = np.where(df['SMOKE'].isin(smoke_enc), df['SMOKE'].map(smoke_enc), np.nan)

nobeysdad_enc = {val:ii for ii, val in enumerate(np.unique(df['NObeyesdad']))}
df['NObeyesdad'] = np.where(df['NObeyesdad'].isin(nobeysdad_enc), df['NObeyesdad'].map(nobeysdad_enc), np.nan)

df
```

```python
#Normalize numeric column into 2 decimal places(age,weight,height)
df['Age'] = df['Age'].apply(lambda x: round(x, 2))
df['Weight'] = df['Weight'].apply(lambda x: round(x, 2))
df['Height'] = df['Height'].apply(lambda x: round(x, 2))
df
```

Find outlier by using Kernel

```python
numeric_columns = ['Age', 'Height', 'Weight']
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(df[column], shade=True)
    plt.title(f'Kernel Density Estimation for {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```
